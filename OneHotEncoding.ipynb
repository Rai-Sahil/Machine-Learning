# Ont Hot Encoding
Fitting Linear Model with Text Data

**Linear Regression Disadvantages**


*   We need to change all the data to numerical values before we send it to train model.
*   It doesn't train on text variables
*   We can use One Hot Encoding Method to convert text variables



**Nominal Categorical Variables**

*   Variables in a data which has no relations between them.
*   Such as (Male, Female), (Delta, Surry, Burnaby) or (Red, Blue, Gree)
---
**Ordinal Categorical Variables**

*   Variables in a data which has  relations between them.
*   Such as (High, Medium, Low), (Satisfied, Unsatisfied, Neutral)



```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
import_data = {
    'Town': ['monroe', 'monroe', 'monroe', 'monroe', 'monroe', 'West Windsor',
             'West Windsor', 'West Windsor', 'West Windsor', 'Robinsville',
             'Robinsville', 'Robinsville', 'Robinsville'],
    'Area': [2600, 3000, 3200, 3600, 4000, 2600, 2800, 3300, 3600, 2600, 2900, 3100, 3600],
    'Price': [550000, 565000, 610000, 680000, 725000, 585000, 615000, 650000, 710000, 575000, 600000, 620000, 695000]
}

df = pd.DataFrame(import_data)
df
```

Create One hot Encoding Datasets from Town Column

```python
dummies = pd.get_dummies(df['Town'])
dummies
```

Merge this above data with original data

```python
merged = pd.concat([df, dummies], axis='columns')
merged
```

Now drop the original town column

```python
merged = merged.drop(['Town'], axis='columns')
merged
```

Now we have to drop one of the dummy columns, it's because we can drive that one dropped variable from rest of the 2 columns of data. Having all 3 can create multi colinear table which can mess up the model.

```python
merged = merged.drop(['West Windsor'], axis='columns')
merged
```

```python
model = LinearRegression()
model.fit(merged[['Area', 'Robinsville', 'monroe']], merged['Price'])
```

```python
# Predict Price in Robinsville
model.predict([[2800, 0, 1]])
```

```python
# Predict Price in West Windsor
model.predict([[3300, 0, 0]])
```

# Verify Model Accuracy

```python
model.score(merged[['Area', 'Robinsville', 'monroe']], merged['Price'])
```
