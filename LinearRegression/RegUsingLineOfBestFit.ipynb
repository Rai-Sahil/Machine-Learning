```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

Create or Import Training Data

```python
years = np.arange(1990, 2021)
income_values = 35000 + 500 * (years - 1990)  # Assuming a linear growth pattern

# Create a DataFrame for Canadian per capita income
df = pd.DataFrame({'Year': years, 'Per Capita Income (CAD)': income_values})
```

Create Plot (Just for us to see)

```python
plt.xlabel('Year')
plt.ylabel('Per Capita Income (CAD)')
plt.scatter(df["Year"], df["Per Capita Income (CAD)"], color='red', marker="+")
```

Have a DataFrame with only year

```python
year = df.drop('Per Capita Income (CAD)', axis="columns")
year
```

Create and Fit the Model

```python
reg = linear_model.LinearRegression()
reg.fit(year, df['Per Capita Income (CAD)'])
```

Make Prediction

```python
reg.predict([[2022]])
```

Import or Create test data

```python
test_data = {'Year': [2022, 2023, 2024, 2025, 2026]}
test_df = pd.DataFrame(test_data)
```

Make Predictions

```python
predicted_price = reg.predict(test_df)
```

```python
test_df['price'] = predicted_price
test_df
```
